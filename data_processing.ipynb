```python
import pandas as pd

data = pd.read_csv('Incendies.csv', sep=';')
data
```

```python
data['Date de première alerte'] = pd.to_datetime(data['Date de première alerte'], format='%Y-%m-%d %H:%M:%S')
data['Date de première alerte']
```

```python
min_date = data['Date de première alerte'].min()
max_date = data['Date de première alerte'].max()

min_date = min_date.date()
print(min_date, max_date)

```
